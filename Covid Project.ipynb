```python
# extractiong key columns
data = df.copy() # backup your data
df = df[["iso_code", "continent","location","date","total_cases","total_deaths", "total_vaccinations"]]
df.head()
```

```python
#exclude_countries = ["Africa","World","Europe","Upper middle income","Western Sahara",
                    #  'Low income','Lower middle income',"High income","Reunion", "Asia","South America", "Oceania"]
exclude_countries = ["Africa","World","Europe", 'Upper middle income', "Western Sahara", 'Low income', 'Lower middle income',
                     "High Income", 'Reunion', "Asia", 'South America', "Oceania",'European Union','North America']
```

```python
# removing the excluded countries
# df.drop(df["location"].isin(exclude_countries)].index)
#df = df[~df["location"].isin(exclude_countries)]
#df["location"].unique()


# removing the excluded countries
# df.drop(df[df["location"].isin(exclude_countries)].index)
df = df[~ df["location"].isin(exclude_countries)]
df["location"].unique()
```

```python
# show total cases and total deaths by countries
cases_per_country = df.groupby("location")[["total_cases","total_deaths"]].sum().sort_values("total_cases",ascending=False).reset_index()
cases_per_country
```

```python
cases_per_country["total_cases"] = cases_per_country["total_cases"].fillna(0)
cases_per_country["total_cases"] 
```

```python
def clean_data(x):
    if x:
        return str(x)[:-3]
    else:
        return 0 
```

```python
cases_per_country["total_cases"] = cases_per_country["total_cases"].apply(clean_data)
cases_per_country
```

```python
cases_per_country["total_cases"] = pd.to_numeric(cases_per_country["total_cases"].replace({'':None,"None":None}), errors='coerce')
cases_per_country["total_cases"] 
```

```python
cases_per_country.columns = ['Country','Total cases', 'Total deaths']
cases_per_countries = cases_per_country.astype({"Country":str,"Total cases":int, "Total deaths":int})
cases_per_countries
```

```python
# converting the date column to datetime
df["date"] = pd.to_datetime(df.date)

# change date format
df["date"] = df["date"].dt.strftime('%y/%m/%d')
df.head()
```

```python
# convert your data to pivote table
data = pd.pivot_table(df,values="total_cases",index=["location"],columns="date").reset_index()
data = data.fillna(0)
data.head()
```

```python
# bring in the flags
flags = pd.read_csv("https://raw.githubusercontent.com/waleolajumoke/flags/main/flags_world.csv")
flags
```

```python
# combining the two data so each country have thier corresponding flags
final= pd.merge(data, flags, left_on="location", right_on="country")
final.head()
```

```python
# swaping the flag column position
del final["country"]
img = final["Image URL"]
del final["Image URL"]
final.insert(1, "Flags", img)
final.head()
```

```python
# export the data for visualization
final.to_csv("covid19result1.csv")
```
